```python
import pandas as pd

trader = pd.read_csv("trader_data.csv")
fg = pd.read_csv("fear_greed.csv")

print(trader.head())
print(trader.columns.tolist())
print(trader.dtypes)
print(fg.head())
```

```python
# Normalise date columns (adjust column names to match your actual data)
trader["date"] = pd.to_datetime(trader["time"]).dt.normalize()
fg["date"]     = pd.to_datetime(fg["Date"]).dt.normalize()

# Merge on date
merged = trader.merge(fg[["date", "Classification", "value"]], on="date", how="left")

# Create a 3-bucket sentiment column
def bucket(c):
    c = str(c).lower()
    if "fear"  in c: return "Fear"
    if "greed" in c: return "Greed"
    return "Neutral"

merged["sentiment"] = merged["Classification"].apply(bucket)
merged["closedPnL"] = pd.to_numeric(merged["closedPnL"], errors="coerce").fillna(0)
```

```python
print("Trader columns:", trader.columns.tolist())
print("Fear/Greed columns:", fg.columns.tolist())
```

```python
import pandas as pd

# ── Auto-detect date column in trader data ──────────────────
date_keywords = ["time", "date", "timestamp", "datetime", "created"]
trader_date_col = next(
    (c for c in trader.columns if any(k in c.lower() for k in date_keywords)),
    None
)
if trader_date_col is None:
    raise ValueError(f"No date column found. Columns are: {trader.columns.tolist()}")
print(f"Using trader date column: '{trader_date_col}'")

# ── Auto-detect date column in fear/greed data ──────────────
fg_date_col = next(
    (c for c in fg.columns if any(k in c.lower() for k in date_keywords)),
    None
)
if fg_date_col is None:
    raise ValueError(f"No date column found. Columns are: {fg.columns.tolist()}")
print(f"Using FGI date column: '{fg_date_col}'")

# ── Auto-detect classification column ───────────────────────
cls_col = next(
    (c for c in fg.columns if any(k in c.lower() for k in ["class", "sentiment", "label", "category"])),
    None
)
print(f"Using classification column: '{cls_col}'")

# ── Auto-detect FGI value column ────────────────────────────
val_col = next(
    (c for c in fg.columns if any(k in c.lower() for k in ["value", "score", "index", "fgi"])),
    None
)
print(f"Using FGI value column: '{val_col}'")

# ── Auto-detect PnL column in trader data ───────────────────
pnl_col = next(
    (c for c in trader.columns if "pnl" in c.lower()),
    None
)
print(f"Using PnL column: '{pnl_col}'")

# ── Normalise dates ─────────────────────────────────────────
trader["date"] = pd.to_datetime(trader[trader_date_col], errors="coerce").dt.normalize()
fg["date"]     = pd.to_datetime(fg[fg_date_col],         errors="coerce").dt.normalize()

# ── Merge ────────────────────────────────────────────────────
cols_to_merge = ["date"]
if cls_col: cols_to_merge.append(cls_col)
if val_col: cols_to_merge.append(val_col)

merged = trader.merge(fg[cols_to_merge], on="date", how="left")

# ── Sentiment bucket ─────────────────────────────────────────
def bucket(c):
    c = str(c).lower()
    if "fear"  in c: return "Fear"
    if "greed" in c: return "Greed"
    return "Neutral"

merged["sentiment"] = merged[cls_col].apply(bucket) if cls_col else "Unknown"

# ── PnL column ───────────────────────────────────────────────
if pnl_col:
    merged["closedPnL"] = pd.to_numeric(merged[pnl_col], errors="coerce").fillna(0)
else:
    raise ValueError(f"No PnL column found. Columns are: {trader.columns.tolist()}")

print(f"\n✅ Merge successful — {len(merged):,} rows")
print(merged[["date", "sentiment", "closedPnL"]].head())
```

```python
# PnL by sentiment
summary = merged.groupby("sentiment")["closedPnL"].agg(
    total_pnl   = "sum",
    mean_pnl    = "mean",
    trade_count = "count",
    win_rate    = lambda x: (x > 0).mean()
)
print(summary)

# Top traders
top_traders = merged.groupby("account")["closedPnL"].agg(
    total_pnl   = "sum",
    win_rate    = lambda x: (x > 0).mean(),
    trade_count = "count"
).sort_values("total_pnl", ascending=False)
print(top_traders.head(20))
```

```python
from scipy import stats

fear_pnl  = merged[merged["sentiment"]=="Fear"]["closedPnL"].dropna()
greed_pnl = merged[merged["sentiment"]=="Greed"]["closedPnL"].dropna()

t_stat, p_val = stats.ttest_ind(fear_pnl, greed_pnl, equal_var=False)
print(f"t={t_stat:.4f}, p={p_val:.6f}")
print("Significant!" if p_val < 0.05 else "Not significant")
```
